```python
# users_di is the user dictionary that store the number of clicks and non-clicks of user, UserID is its key
users_di = {}
def count_user_clicks(x):
    if not users_di.has_key(x['UserID']):
        users_di[x['UserID']] = {'clicks_no': 0 , 'non_clicks_no': 0 }
    if x['IsClick']==1:
        users_di[x['UserID']]['clicks_no'] += 1 
    else:
        users_di[x['UserID']]['non_clicks_no'] += 1
```

```python
#read the dataframe by chunk and apply the counting function
for chunk in pd.read_csv(SOURCE, sep=',' , chunksize=100000):
    #print len(users_di)
    chunk.apply(count_user_clicks , axis =1 )
```

```python
# search_di is the search query dictionary that store the number of regular, highlighted, context ads, SearchID is its key
search_di = {}
def count_search_ads(x):
    if not search_di.has_key(x['SearchID']):
        search_di[x['SearchID']] = {'regular_no': 0 , 'highlighted_no': 0 , 'context_no': 0 }
    if x['ObjectType']==1:
        search_di[x['SearchID']]['regular_no'] += 1 
    elif x['ObjectType']==2:
        search_di[x['SearchID']]['highlighted_no'] += 1 
    else:
        search_di[x['SearchID']]['context_no'] += 1
```

```python
#read the dataframe by chunk and apply the counting function
SOURCE2 = "../data/testSearchStream.tsv"
for chunk in pd.read_csv(SOURCE2, sep='\t', nrows=100000 , chunksize=100000):
    #print len(search_di)
    chunk.apply(count_search_ads , axis =1 )
```
